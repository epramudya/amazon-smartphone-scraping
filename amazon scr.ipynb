```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def scrape_amazon_page(url, headers):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    products = soup.find_all('div', {'data-component-type': 's-search-result'})
    
    product_names = []
    product_prices = []
    product_ratings = []
    product_reviews = []

    for product in products:
        # Mendapatkan nama produk
        name = product.h2.text.strip()
        product_names.append(name)

        # Mendapatkan harga produk
        price_whole = product.find('span', class_='a-price-whole')
        price_fraction = product.find('span', class_='a-price-fraction')
        if price_whole and price_fraction:
            price = f"${price_whole.text.strip()}.{price_fraction.text.strip()}"
        else:
            price = "N/A"
        product_prices.append(price)

        # Mendapatkan rating produk
        rating = product.find('span', {'class': 'a-icon-alt'})
        if rating:
            product_ratings.append(rating.text.strip())
        else:
            product_ratings.append("N/A")

        # Mendapatkan jumlah penilaian produk
        reviews = product.find('span', {'class': 'a-size-base'})
        if reviews:
            product_reviews.append(reviews.text.strip())
        else:
            product_reviews.append("N/A")

    return product_names, product_prices, product_ratings, product_reviews
```

MENCARI NAMA PRODUCT

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36",
    "Accept-Language": "en-US, en;q=0.5"
}

# Loop untuk scraping hingga halaman ke-3
base_url = 'https://www.amazon.com/s?k=smartphones'
all_products = []

for page in range(1, 4):  # Mulai dari halaman 1 hingga halaman 3
    url = f'{base_url}&page={page}'
    products, _, _, _ = scrape_amazon_page(url, headers)
    all_products.extend(products)

# Menampilkan output dalam bentuk tabel pandas
df = pd.DataFrame({'Nama Produk': all_products})
df
```

MENCARI HARGA PRODUCT

```python
all_products = []
all_prices = []

for page in range(1, 4):  # Mulai dari halaman 1 hingga halaman 3
    url = f'{base_url}&page={page}'
    products, prices, _, _ = scrape_amazon_page(url, headers)
    all_products.extend(products)
    all_prices.extend(prices)

# Menampilkan output dalam bentuk tabel pandas
df = pd.DataFrame({'Nama Produk': all_products, 'Harga': all_prices})
df
```

MENCARI RATING DAN TOTAL PENILAIAN PRODUCT

```python
all_products = []
all_prices = []
all_ratings = []
all_reviews = []

for page in range(1, 4):  # Mulai dari halaman 1 hingga halaman 3
    url = f'{base_url}&page={page}'
    products, prices, ratings, reviews = scrape_amazon_page(url, headers)
    all_products.extend(products)
    all_prices.extend(prices)
    all_ratings.extend(ratings)
    all_reviews.extend(reviews)

# Menampilkan output dalam bentuk tabel pandas
df = pd.DataFrame({
    'Nama Produk': all_products,
    'Harga': all_prices,
    'Rating': all_ratings,
    'Jumlah Penilaian': all_reviews
})
df
```
